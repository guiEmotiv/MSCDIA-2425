# **Sobre este archivo**
Consta de 25 métricas y 40mil registros. Las columnas son las siguientes:

* Timestamp: La hora en la que ocurrió la actividad de la red.
* Source IP Address: La dirección IP del remitente o iniciador del tráfico de red.
* Destination IP Address: La dirección IP del receptor o objetivo del tráfico de red.
* Source Port: El número de puerto utilizado por la dirección IP de origen.
* Destination Port: El número de puerto utilizado por la dirección IP de destino.
* Protocol: El protocolo de comunicación utilizado (por ejemplo, TCP, UDP, ICMP).
* Packet Length: El tamaño del paquete en bytes.
* Packet Type: Tipo de paquete (por ejemplo, paquete de datos, paquete de control).
* Traffic Type: El tipo de tráfico (por ejemplo, tráfico web, tráfico de correo electrónico).
* Payload Data: Los datos reales transmitidos en el paquete.
* Malware Indicators: Indicadores de actividad potencialmente maliciosa o presencia de malware.
* Anomaly Scores: Puntajes que indican desviaciones del comportamiento esperado, utilizados para la detección de anomalías.
* Alerts/Warnings: Notificaciones o advertencias generadas por sistemas de seguridad o herramientas de monitoreo.
* Attack Type: Tipo de ataque detectado o sospechado (por ejemplo, DDoS, inyección SQL).
* Attack Signature: Patrones o firmas específicos asociados con ataques conocidos.
* Action Taken: Acciones realizadas en respuesta a amenazas o anomalías detectadas.
* Severity Level: El nivel de severidad asociado con una alerta o evento (por ejemplo, bajo, medio, alto).
* User Information: Información sobre el usuario involucrado en la actividad de la red.
* Device Information: Información sobre el dispositivo involucrado en la actividad de la red (por ejemplo, tipo de dispositivo, sistema operativo).
* Network Segment: El segmento o subred de la red donde ocurrió la actividad.
* Geo-location Data: Información de ubicación geográfica asociada con direcciones IP.
* Proxy Information: Información sobre los servidores proxy involucrados en la comunicación de la red.
* Firewall Logs: Registros generados por dispositivos de firewall que indican tráfico permitido o bloqueado.
* IDS/IPS Alerts: Alertas generadas por Sistemas de Detección de Intrusos (IDS) o Sistemas de Prevención de Intrusos (IPS) que indican actividad sospechosa o maliciosa.
* Log Source: La fuente u origen de la entrada de registro (por ejemplo, nombre del sistema o dispositivo de registro).

````python
# @title 15. Crea la columna 'Browser'. Esta columna contiene solo el explorador encontrado en 'Device Information'  (1/2pt)
# Ejm: Valor de 'Device Information': 'Mozilla/5.0 (compatible; MSIE 6.0; Windows NT 6.2; Trident/3.0)' , valor de 'Browser': 'Mozilla'
# Agrega tu código a partir de la siguiente línea
def extraer_browser(device_information):
    return device_information.split('/')[0]
df['Browser'] = df['Device Information'].apply(extraer_browser)
df.head(20)
````

````python
# NO MODIFIQUE ESTE BLOQUE DE CODIGO, SOLO EJECUTELO

import re

patterns = [
    r'Windows',
    r'Linux',
    r'Android',
    r'iPad',
    r'iPod',
    r'iPhone',
    r'Macintosh',
]

def extract_device_or_os(user_agent):
    for pattern in patterns:
        match = re.search(pattern, user_agent, re.I)
        if match:
            return match.group()
    return 'Unknown'
````

````python
# @title 16. Crea la columna 'Device/OS'. Aplicando la función extract_device_or_os a la columna 'Device Information'  (1/2pt)
# TIP: Usa la función 'apply'
# Agrega tu código a partir de la siguiente línea
df['Device/OS'] = df['Device Information'].apply(extract_device_or_os)
df.head(20)
````

````python
# @title 17. Cuente los valores existentes en las columnas: 'Browser' y 'Device/OS'  (1/2pt)
# Agrega tu código a partir de la siguiente línea
print(df.value_counts('Device/OS'))
print(df.value_counts('Browser'))
````

````python
# @title 18. Elimine la columna 'Device Information'  (1/2pt)
# Agrega tu código a partir de la siguiente línea
df.drop('Device Information', axis=1, inplace=True)
df.head(20)
````

````python
# @title 19. Genera las columnas: 'Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'DayOfWeek' a partir de la columna 'Timestamp'  (1pt)
# TIP: Puedes usar la función apply y crear una función para esto
# Agrega tu código a partir de la siguiente línea
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
def extract_timestamp(ts):
    return pd.Series([ts.year, ts.month, ts.day, ts.hour, ts.minute, ts.second, ts.dayofweek])
df[['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'DayOfWeek']] = df['Timestamp'].apply(extract_timestamp)
df.head(20)
````
